```python
import cv2
import numpy as np
from ultralytics import YOLO
from matplotlib import pyplot as plt

def calculate_and_visualize_with_correct_green_area(image_path, model):
    # Run inference
    results = model.predict(source=image_path, conf=0.25, retina_masks=True)

    # Ensure masks are available
    if not results[0].masks:
        raise ValueError("No masks detected in the image!")

    # Extract masks and class IDs
    masks = results[0].masks.data.cpu().numpy()  # Segmentation masks
    class_ids = results[0].boxes.cls.cpu().numpy()  # Corresponding class IDs

    # Define class IDs based on your dataset
    BUILDING_ID = 0
    ROAD_ID = 1
    WHITE_SPACE_ID = 2

    # Load the original image
    original_image = cv2.imread(image_path)
    original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert to RGB for visualization

    # Create a combined mask for visualization
    combined_mask = np.zeros_like(original_image, dtype=np.uint8)

    # Overlay masks for each class with different colors
    for mask, cid in zip(masks, class_ids):
        color = (0, 255, 0) if cid == BUILDING_ID else (255, 0, 0) if cid == ROAD_ID else (0, 0, 255)
        combined_mask[mask > 0] = color

    # Calculate areas for each class
    image_area = original_image.shape[0] * original_image.shape[1]  # Total area in the image
    building_area = sum(mask.sum() for mask, cid in zip(masks, class_ids) if cid == BUILDING_ID)
    road_area = sum(mask.sum() for mask, cid in zip(masks, class_ids) if cid == ROAD_ID)
    white_space_area = sum(mask.sum() for mask, cid in zip(masks, class_ids) if cid == WHITE_SPACE_ID)

    # Calculate green area as the remaining background
    total_masked_area = building_area + road_area + white_space_area
    green_area = image_area - total_masked_area

    # Avoid division by zero
    total_area = max(image_area, 1)

    # Calculate percentages
    green_percentage = (green_area / total_area) * 100
    concrete_percentage = ((building_area + road_area) / total_area) * 100

    # Visualization
    plt.figure(figsize=(15, 10))

    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(original_image)
    plt.title("Original Image")
    plt.axis("off")

    # Combined mask
    plt.subplot(1, 3, 2)
    plt.imshow(combined_mask)
    plt.title("Masked Image")
    plt.axis("off")

    # Overlay of original and masks
    overlay = cv2.addWeighted(original_image, 0.7, combined_mask, 0.3, 0)
    plt.subplot(1, 3, 3)
    plt.imshow(overlay)
    plt.title("Overlay Image")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    # Print detailed results
    details = {
        "Total Area (pixels)": total_area,
        "Building Area (pixels)": building_area,
        "Road Area (pixels)": road_area,
        "White Space Area (pixels)": white_space_area,
        "Green Area (pixels)": green_area,
    }

    return {
        "Natural Area Percentage": green_percentage,
        "Concrete Area Percentage": concrete_percentage,
        "Details": details,
    }

# Usage example
image_path = "C:\\Users\\Dharani\\Desktop\\jupyter\\yolo_withoutmain\\yolo_withoutmain\\Green-to-Concrete-Ratio-5\\train\\images\\pic40_png.rf.aa71fe4e7407dc2d506e640af581fc31.jpg"# Replace with your test image path
model_path = "C:\\Users\\Dharani\\Desktop\\jupyter\\yolo_withoutmain\\yolo_withoutmain\\runs\\segment\\train4\\weights\\best.pt"  # Replace with your trained model path
model = YOLO(model_path)

# Calculate ratios and visualize
ratios = calculate_and_visualize_with_correct_green_area(image_path, model)

# Print results
print(f"Natural Area Percentage: {ratios['Natural Area Percentage']:.2f}%")
print(f"Concrete Area Percentage: {ratios['Concrete Area Percentage']:.2f}%")

print("\nDetails:")
for key, value in ratios["Details"].items():
    print(f"{key}: {value}")
```
